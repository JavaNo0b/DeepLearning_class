# CSV 파일 불러오기

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

df = pd.read_csv("C:/Users/82107/Desktop/deepLearning/week1/abalone.csv", index_col=0)
df
```

```python
# 컬렴명 확인
df.columns
```

```python
df.isnull().sum()
df = df.dropna()
df.isnull().sum()
```

```python
# Rings
sns.countplot(data = df, x="Rings")
plt.xlabel("Rings")
plt.ylabel("Count")
plt.title("Rings")
```

```python
df['Rings'].value_counts()
```

```python
# 원핫 인코딩 (Sex 열만)
df_sex_encoded = pd.get_dummies(df['Sex'], prefix='Sex')

# 숫자형으로 변환
df_sex_encoded = df_sex_encoded.astype(int)

# 원래 데이터프레임과 원핫 인코딩된 데이터프레임 합치기
df_final = pd.concat([df.drop('Sex', axis=1), df_sex_encoded], axis=1)
```

```python
df_final
```

```python
X=df_final.drop('Rings',axis=1)
X.head()
```

```python
# LabelEncoder : 라벨 인코딩 = 이 객체는 범주형 데이터를 숫자로 변환하는 데 사용 / [7, 9, 10, 15]라면, 라벨 인코딩 후 [0, 1, 2, 3]으로 변환
encoder = LabelEncoder()
y=df_final['Rings']
y_encoded = encoder.fit_transform(y)
```

```python
pd.Series(y_encoded).value_counts()
```

```python
print ("==================================")
print (" ACCURACY OF OTHER ML CLASSIFIERS" )
print ("==================================")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded,test_size=0.20,random_state=0) # train데이터와 test데이터로 나눔
# test_size=0.25 : test 를 20%, train을 80%로 나눔
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
#SVM
from sklearn.svm import SVC

clf_svm = SVC(random_state=0) # 모델 만들고
clf_svm.fit(X_train, y_train) # train 데이터로 학습 시키고

pred_svm = clf_svm.predict(X_test) # X_test로 예측하고

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm)) # 예측한 값과 y_test값과 비교하기
print(confusion_matrix(y_test, pred_svm)) # 성능 평가
```

```python
# Build a logistic regression classifier and predict
#LR

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
# Build a decision tree classifier and predict
#DT

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(X_train, y_train)

pred_dt = clf_dt.predict(X_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
#RF
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print ("\n--- Radom Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print("\n--- KNN ---")
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_train)
pred = knn_clf.predict(X_test)
print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
```
