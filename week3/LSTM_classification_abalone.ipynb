```python
from sklearn.preprocessing import LabelEncoder
# LabelEncoder 객체 생성 후 라벨인코딩 수행
label_encoder = LabelEncoder()
df['Rings'] = label_encoder.fit_transform(df['Rings'])
```

```python
df.isnull().sum()
df = df.dropna()
df.isnull().sum()
```

```python
# 원핫 인코딩 (Sex 열만)
df_sex_encoded = pd.get_dummies(df['Sex'], prefix='Sex')

# 숫자형으로 변환
df_sex_encoded = df_sex_encoded.astype(int)

# 원래 데이터프레임과 원핫 인코딩된 데이터프레임 합치기
df = pd.concat([df.drop('Sex', axis=1), df_sex_encoded], axis=1)
```

```python
X=df.drop('Rings',axis=1)
X.head()
```

```python
y=df['Rings']
y.value_counts()
```

```python
# 각 컬럼별 히스토그램을 그리는 코드
df.hist(figsize=(15, 20), bins=30)  # ax 인수 없이 자동으로 서브플롯 생성
plt.tight_layout()  # 서브플롯 간격을 자동으로 조정
plt.show()
```

```python
X = X.values
X.shape
```

```python
# 0,1 로 표현하는 원핫인코딩
from tensorflow.keras.utils import to_categorical

Y = to_categorical(y)
Y
```

```python
print(X.dtype)
print(Y.dtype)
```

```python
# 스케일링
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 윈도우 크기를 설정하고, 데이터를 3차원으로 변환 (타임스텝 적용)
def create_windows(X, y, window_size):
    X_windows = []
    y_windows = []
    for i in range(len(X) - window_size):
        X_windows.append(X[i:i+window_size])  # window_size만큼의 과거 데이터를 사용
        y_windows.append(y[i+window_size])    # 그 다음 값을 예측
    return np.array(X_windows), np.array(y_windows)
```

```python
window_size = 3  # 3개의 타임스텝을 사용
X_windows, Y_windows = create_windows(X_scaled, Y, window_size)
print (X_windows)
print (Y_windows)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_windows, Y_windows, test_size=0.2, random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout,LSTM

# LSTM 모델 구성
# Sequential 모델 : 레이어를 순차적으로 쌓아올리는 방식으로 신경망을 구성
# Dropout 레이어를 추가하여 과적합을 방지. 드롭아웃 비율 0.2는 20%의 뉴런을 무작위로 비활성화
# Dense 레이어는 LSTM의 출력을 고차원 공간으로 매핑
# 타임스텝만큼 과거의 연속적인 데이터를 처리
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(3, 10), return_sequences=True))  # 타임스텝=3, 특성=10
model.add(Dropout(0.2))  # Dropout 레이어 추가
model.add(LSTM(32, activation='relu'))  # 두 번째 LSTM 레이어
model.add(Dropout(0.2))  # Dropout 레이어 추가
model.add(Dense(64, activation='relu'))  # Fully Connected Layer
model.add(Dropout(0.5))  # Dropout 레이어 추가
model.add(Dense(28, activation='softmax'))  # 출력층: 28개 클래스

# 모델 컴파일
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 모델 학습
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test,axis=1)
y_pred_class = np.argmax(y_pred,axis=1)
```

```python
# Loss
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# 정확도
plt.plot(history.history['accuracy'], 'b', label='Training ACCURACY')
plt.plot(history.history['val_accuracy'], 'y', label='Validation ACCURACY')
plt.title('Training and Validation ACCURACY')
plt.xlabel('Epochs')
plt.ylabel('ACCURACY')
plt.legend()
plt.show()
```

```python
#Accuracy of the predicted values
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_class,y_pred_class, zero_division=0))
print(confusion_matrix(y_test_class,y_pred_class))
```
